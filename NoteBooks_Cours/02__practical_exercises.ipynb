## Exercise 2

- Using the reduced numerical features obtained with PCA in the practical
  exercise `01__practical_exercises.ipynb`.
- Then, apply MDS, LLE, Isomap, or TSNE.

```python
categorical_vars = ['Suburb', 'Type', 'Method', 'SellerG', 'CouncilArea', 'Regionname', "Address", "CouncilArea_mode", "Date"]
numerical_vars = list(set(melb_df.columns) - set(categorical_vars))

melb_df_num = (
    (melb_df[numerical_vars] - melb_df[numerical_vars].mean(axis=0)
) / melb_df[numerical_vars].std(axis=0))

from sklearn.decomposition import PCA
pca = PCA(n_components=4)
pca.fit(melb_df_num)
proyected_features = pca.transform(melb_df_num)
```
